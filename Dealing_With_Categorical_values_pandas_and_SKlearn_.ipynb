# Dealing With Categorical Values

## Importing Libraries

```python
import pandas as pd
import numpy as np

```

## Importing Dataset

```python
df_1 = pd.read_csv("Salary_Dataset.csv")
```

```python
df_1.head(40)
```

# Lets perform Encoding

## Applying One-Hot Encoding

<h3> First Do it with pandas </h3>

```python
# create dummy variables for the "Gender" column
country_dummies = pd.get_dummies(df_1['country'])

country_dummies
```

```python
# concatenate the dataframes into original dataframes.
df1 = pd.concat([df_1,country_dummies], axis=1)
```

```python
df1
```

```python
df1.drop('country', axis =1, inplace = True)
```

```python
df1.head()
```

```python
# reaarrange the columns
df2 = df1[['Australia',	'Canada',	'Dubai',	'USA', 'Salary',	'YearsExperience',	'Purchased'	]]
df2
```

```python
df2.info()
```

```python
# convert the "Salary" column from float to int and save to the original dataframe
df2['Salary'] = df2['Salary'].astype('int')


# print the updated dataframe
df2.head()
```

```python
df2.to_csv('encoded_salary_dataset.csv')
```

# <h3> Let's do it with Scikit-Learn </h3>
Also Restart your kernel if you are using same notebook


```python
df_1 = pd.read_csv("Salary_Dataset.csv")
df_1.head(1)
```

```python
# Lets first perform label encoding
# Labelencoding will be performed on Purchased column as it has only two unique value in it.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_1["Purchased"] = le.fit_transform(df_1['Purchased'])
df_1.head()
```

```python
# Lets perform one-hot encoding on country column.
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer (transformers=[('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
df_1 = pd.DataFrame(ct.fit_transform(df_1))
```

```python
#Lets link the headers name.
df_1.columns =['country','Australia',	'Canada',	'Dubai',	'USA', 'Salary',	'YearsExperience',	'Purchased'	]
```

```python
df_1.head()
```

```python
df_1.drop('country', axis=1, inplace=True)
```

```python
df_1.head()
```

```python
df_1= df_1.to_csv('prep_slary_daatset')
```
